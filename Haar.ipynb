```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
#importing time library for speed comparisons of both classifiers
import time 
%matplotlib inline
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#load test iamge
test1 = cv2.imread('.\images\kid.jpg')

#test2 = convertToRGB(test1)
#print(test1)
#convert the test image to gray image as opencv face detector expects gray images
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
```

```python
#if you have matplotlib installed then 
plt.imshow(gray_img, cmap='gray')

# or display the gray image using OpenCV
# cv2.imshow('Test Imag', gray_img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
path = 'C:\Anaconda3\envs\myWindowsCV\Library\etc\haarcascades\haarcascade_frontalface_default.xml'
#load cascade classifier training file for haarcascade#load ca 
haar_face_cascade = cv2.CascadeClassifier(path)
```

```python

#let's detect multiscale (some images may be closer to camera than others) images#let's  
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);

#print the number of faces found
print('Faces found: ', len(faces))
```

```python
#go over list of faces and draw them as rectangles on original colored img
for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
```
